# Seaborn Basics

```python
# Import Seaborn
import seaborn as sns
```

```python
# Load demo databaset
tips = sns.load_dataset("tips")

# Documents
# https://github.com/mwaskom/seaborn-data
```

```python
tips.head()
```

```python
# Bar Plot
sns.barplot(tips, x=tips["tip"], y=tips["sex"], color="green")

# Documentation
# https://seaborn.pydata.org/generated/seaborn.barplot.html#codecell0
```

```python
import matplotlib.pyplot as plt
```

```python
# Scatter Plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="time").grid(visible=False)
plt.title("Total bill vs Tip")
plt.show()

# Documentation
# https://seaborn.pydata.org/generated/seaborn.scatterplot.html#codecell0
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
# Line Plot
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers").grid(visible=False)

# Documentation
# https://seaborn.pydata.org/generated/seaborn.lineplot.html#codecell0
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month").set_title("Passengers by Year").set_label()
```

```python
titanic = sns.load_dataset("titanic")
```

```python
# Box Plot
sns.boxplot(data=titanic, x="class", y="age", hue="alive")

# Documentation
# https://seaborn.pydata.org/generated/seaborn.boxplot.html#codecell2
```

```python
# Violin Plot
# sns.violinplot(x=titanic["age"])
sns.violinplot(data=titanic, x="age", y="class", hue="alive")

# Documentation
# https://seaborn.pydata.org/generated/seaborn.violinplot.html#codecell2
```

```python
# Histogram Plot
penguins = sns.load_dataset('penguins')
sns.histplot(data=penguins, x="flipper_length_mm", binwidth=3, kde=True).set_xlabel("Flipper Length MM")

# Documentation
# https://seaborn.pydata.org/generated/seaborn.histplot.html#codecell0
```

```python
# KDE (Kernel density estimate) Plot
sns.kdeplot(data=tips, x="total_bill", fill=True)

# Documentation
# https://seaborn.pydata.org/generated/seaborn.kdeplot.html#codecell0
```

```python
# Pair Plot
sns.pairplot(penguins, hue="species", markers=["o", "s", "D"])

# Documentation
# https://seaborn.pydata.org/generated/seaborn.pairplot.html#codecell5
```

```python
# Heat Map
glue = sns.load_dataset("glue")
pivot = glue.pivot(index="Model", columns="Task", values="Score")
sns.heatmap(pivot, annot=True)

# Documentation
# https://seaborn.pydata.org/generated/seaborn.heatmap.html#codecell0
```
